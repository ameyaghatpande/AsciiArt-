#### This code will create an AsciiArt Encoder/Decoder

#### Data format - I am using Run Length Encoding(RLE) to compress the input Ascii Art. This encoding is useful to compress the input data when it contains "CONSECUTIVE" similar characters. I choose this compression type becuase Ascii Art has lot of consecutive characters and also has lot of tabs/spaces at start/end of each line (which is again a character " "). This helps to compress the data effeciently
#### Sample input = "AAA2222CCCCCCA"  
#### Sample encoded output by RLE = [(3, A), (4, 2), (6, C), (1,A)]

#### After encoding the AsciiArt, I am not saving the encoded format onto disk. I am using a class variable to store the encoded format.In real world application, this can be saved on the disk

```python
# <aghatpande> on 20-Jun-2017 for BetaBrand test
import sys
from itertools import groupby

class AsciiArt():
    
    # Initializing a list to hold the encoded AsciiArt
    encodedAsciiArt = []
  
    # <aghatpande> on 20-Jun-2017 for BetaBrand test
    # I have "Data.txt" in the same folder from where I am running this jupyter file
    # Need to give the absolute path, if your Data.txt is somewhere else.
    def __init__(self, fileName):    

        self.inputFile = open(fileName, "r")

        # Enable the below line of code for debugging
        # print inputFile.read()
        
    # <aghatpande> on 20-Jun-2017 for BetaBrand test      
    # encodeIntoRLE function will use groupby to count each consecutive similar characters
    # Sample input = "AAA2222CCCCCCA"
    # Sample output = [(3, A), (4, 2), (6, C), (1,A)]
    def encodeIntoRLE(self, inputLine):
        return [(len(list(groups)), character) for character,groups in groupby(inputLine)]

    # <aghatpande> on 20-Jun-2017 for BetaBrand test
    # Reading the ASCII ART per line , and appending the RLE format in a list
    # The encoded format will have "\n", so no extra efforts needed while decoding
    def encodeAsciiArt(self):
        lines = self.inputFile.readlines()
        for line in lines:
            self.encodedAsciiArt.append(self.encodeIntoRLE(line))   
 
    # <aghatpande> on 20-Jun-2017 for BetaBrand test
    # Helper function to see the output of Encoded string
    def printEncodedAsciiArt(self):
        if self.encodedAsciiArt:
            print self.encodedAsciiArt

    # <aghatpande> on 20-Jun-2017 for BetaBrand test
    # decodeFromRLE function will decode the input string
    # Expected input is in form of tuples. Each tuple will have (Occurence_count , character)
    # The input tuples will also have one of the entries as "\n", which is required to maintain the shape of ASCII ART 
    def decodeFromRLE(self, inputTuple):
        return ''.join(character * count for count,character in inputTuple)

    # <aghatpande> on 20-Jun-2017 for BetaBrand test
    # Decode to restore the AsciiArt. Here "sys.stdout.write" is used instead of "print" function 
    # since print inserts extra "\n" by default which is not reuqired
    def decodeAsciiArt(self):
        for items in self.encodedAsciiArt:
            sys.stdout.write(self.decodeFromRLE(items))
            
        # Close the fileObject instance which is created in constructor   
        self.inputFile.close() 
 
    # <aghatpande> on 20-Jun-2017 for BetaBrand test
    # Close the file explicitly if the above function is never called.
    def closeInputFile(self):
        if self.inputFile:
            self.inputFile.close()
            self.inputFile = None
        else:
            print("File is already closed")
 
    ############################## All the below functions are used for Unit testing ##############################

    # <aghatpande> on 20-Jun-2017 for BetaBrand test
    # checkInputFile is used for Unit testing
    # Check if the input File exists
    # Return = True if File not empty Else False
    def checkInputFile(self):
        if self.inputFile:
            return True
        else:
            return False
        
    # <aghatpande> on 20-Jun-2017 for BetaBrand test
    # checkAsciiEncoding is used for Unit testing
    # Check if the input File has been succesfully encoded (File not empty)
    # Return = True if size of encoding is > 0 else False
    def checkAsciiEncoding(self):
        if self.inputFile:
            if len(self.encodedAsciiArt) != 0:
                return True
            else:
                return False
        else:
            return False
        
    # <aghatpande> on 20-Jun-2017 for BetaBrand test
    # checkNonAsciiChars is used for Unit testing
    # Check if the input File has any Non-Ascii characters
    # Return = True if Non Ascii characters is present else False
    def checkNonAsciiChars(self):
        if self.inputFile:
            for items in self.encodedAsciiArt:
                for count,character in items:
                    if(ord(character) > 255):
                        return True    
            return False
        else:
            return False
            
    # <aghatpande> on 20-Jun-2017 for BetaBrand test 
    # checkBadDecoding is used for Unit testing
    # Check while decoding , no character has 0 occurences. Any character should have atleast 1 occurence 
    # Return = True if character has 0 occurence else False
    def checkBadDecoding(self):
        if self.inputFile:
            for items in self.encodedAsciiArt:
                for count,character in items:
                    if (int(count) == 0):
                        return True
            return False
        else:
            return False
```

```python
# <aghatpande> on 20-Jun-2017 for BetaBrand test
# Load the file
firstObject = AsciiArt("Data.txt")

# Encode the AsciiArt
firstObject.encodeAsciiArt()

# Enable the below line of code for testing
# firstObject.printEncodedAsciiArt()

#Decode the AsciiArt
firstObject.decodeAsciiArt()

#firstObject.closeInputFile()
```
